```python
import numpy as np
import pandas as pd
import random

import matplotlib.pyplot as plt
import seaborn as sns

from kaggle_environments import make, evaluate
```

```python
import random

def rock_agent(observation, configuration):
    """
    Возвращает 0, всегда выбирает камень.
    """
    return 0

def paper_agent(observation, configuration):
    """
    Возвращает 1, всегда выбирает бумагу.
    """
    return 1

def scissors_agent(observation, configuration):
    """
    Возвращает 2, всегда выбирает ножницы.
    """
    return 2

def random_agent(observation, configuration):
    """
    Возвращает случайный выбор из каменя, ножницы или бумаги.
    """
    return random.choice([0, 1, 2])

def copy_opponent_agent(observation, configuration):
    """
    Возвращает действие последнего оппонента, иначе случайный выбор.
    """
    if observation.get('step', 0) > 0:
        return observation.get('lastOpponentAction', random.choice([0, 1, 2]))
    else:
        return random.choice([0, 1, 2])

def beat_last_agent(observation, configuration):
    """
    Возвращает действие, которое бьет последнего оппонента, иначе случайный выбор.
    """
    if observation.get('step', 0) > 0:
        return (observation.get('lastOpponentAction', 0) + 1) % 3
    else:
        return random.choice([0, 1, 2])

def cycle_agent(observation, configuration):
    """
    Возвращает 0, иначе возвращает действие, следующее за последним действием оппонента.
    """
    if observation.get('step', 0) == 0:
        return 0
    else:
        return (observation.get('lastOpponentAction', 0) + 1) % 3

def counter_cycle_agent(observation, configuration):
    """
    Возвращает 0, иначе возвращает действие, следующее за последним действием оппонента с шагом 2.
    """
    if observation.get('step', 0) == 0:
        return 0
    else:
        return (observation.get('lastOpponentAction', 0) + 2) % 3

def frequency_agent(observation, configuration):
    """
    Вычисляет и возвращает наиболее частое действие в истории.
    """
    if observation.get('step', 0) > 0:
        counts = [0, 0, 0]
        for action in observation.get('history', []):
            counts[action] += 1
        return counts.index(max(counts))
    else:
        return random.choice([0, 1, 2])

def beat_frequency_agent(observation, configuration):
    """
    Вычисляет и возвращает действие, которое бьет наиболее частое действие в истории.
    """
    if observation.get('step', 0) > 0:
        counts = [0, 0, 0]
        for action in observation.get('history', []):
            counts[action] += 1
        return (counts.index(max(counts)) + 1) % 3
    else:
        return random.choice([0, 1, 2])

def mixed_strategy_agent_rock(observation, configuration):
    """
    Смешанная стратегия с преимуществом для камня (0).
    """
    mixed_choices = [0, 0, 0, 1, 1, 2]
    return random.choice(mixed_choices)

def mixed_strategy_agent_paper(observation, configuration):
    """
    Смешанная стратегия с преимуществом для бумаги (1).
    """
    mixed_choices = [0, 0, 1, 1, 1, 2]
    return random.choice(mixed_choices)

def mixed_strategy_agent_scissors(observation, configuration):
    """
    Смешанная стратегия с преимуществом для ножниц (2).
    """
    mixed_choices = [0, 1, 1, 2, 2, 2]
    return random.choice(mixed_choices)

def adaptive_agent(observation, configuration):
    if observation.get('step', 0) > 1:
        if observation.get('lastOpponentAction', 0) == observation.get('history', [0, 0])[-2]:
            return (observation.get('lastOpponentAction', 0) + 1) % 3
        else:
            return 0
    else:
        return 0
```

```python
# Список агентов
agents = [
    rock_agent,
    paper_agent,
    scissors_agent,
    random_agent,
    copy_opponent_agent,
    beat_last_agent,
    cycle_agent,
    counter_cycle_agent,
    frequency_agent,
    beat_frequency_agent,
    mixed_strategy_agent_paper,
    mixed_strategy_agent_rock,
    mixed_strategy_agent_scissors,
    adaptive_agent
]

# Словарь для хранения очков каждой стратегии
rank = {agent: 0 for agent in agents}

# Цикл для проведения 3 раундов
for times in range(3):
    # Сколько раундов сыграет каждая пара
    for i in range(len(agents)):
        for j in range(i + 1, len(agents)):
            # Оценка производительности стратегий
            e = evaluate(
                "rps",  # Среда для использования - не нужно менять
                [agents[i], agents[j]],
                configuration={"episodeSteps": 100, 'tieRewardThreshold': 1}
            )
            # Сравнение результатов
            if e[0][0] > e[0][1]:
                # Если первая стратегия выиграла, прибавляем ему очко
                rank[agents[i]] += 1
            elif e[0][0] < e[0][1]:
                # Если вторая стратегия выиграла, увеличиваем ему очко
                rank[agents[j]] += 1
```

```python
# Создание рейтинга на основе результатов стратегий
rank_se = pd.Series(rank, name="Рейтинг турнира по количеству выигранных партий")

# Сортировка значений по убыванию
rank_se = rank_se.sort_values(ascending=False)

# Преобразование серии в фрейм данных
rank_se = rank_se.to_frame()

# Вывод фрейма данных
rank_se
```
